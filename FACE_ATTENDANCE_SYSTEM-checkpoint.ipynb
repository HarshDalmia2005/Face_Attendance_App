```python
from tkinter import *
import tkinter as tk
import customtkinter as ck
from PIL import ImageTk,Image
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
import time
from tkinter import messagebox
from tkinter import ttk
import cv2
import os


class face_recog:
    def __init__(self,root):
        self.recog()
        if not firebase_admin._apps:
            cred = credentials.Certificate("serviceaccountkey.json")
            firebase_admin.initialize_app(cred, {
                'databaseURL': "https://faceattendance-63040-default-rtdb.firebaseio.com/"
            })
        
    def recog(self):
        # Load the trained model
        recognizer = cv2.face.LBPHFaceRecognizer_create()
        recognizer.read('face_recognizer.yml')

        # Load the label map
        label_map = {}
        with open('label_map.txt', 'r') as f:
            for line in f:
                label, name = line.strip().split(',')
                label_map[int(label)] = name

        # Function to recognize faces from webcam
        def recognize_faces_from_webcam():
            
            if not firebase_admin._apps:
                cred = credentials.Certificate("serviceaccountkey.json")
                firebase_admin.initialize_app(cred, {
                    'databaseURL': "https://faceattendance-63040-default-rtdb.firebaseio.com/"
                })
            
            ref=db.reference('Students')
            list=ref.get()
            # Open the webcam
            cap = cv2.VideoCapture(0)
            frame=cv2.imread("background.png")
            cap.set(3,640)
            cap.set(4,480)
            folder='Modes'
            modepathlist=os.listdir(folder)
            imgmodelist=[]
            counter=0
            modetype=0
            
            for path in modepathlist:
                imgmodelist.append(cv2.imread(os.path.join(folder,path)))
            
            if not cap.isOpened():
                print("Error: Could not open webcam.")
                return

            face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

            while True:
                ret, img = cap.read()
                frame[162:162+480,55:55+640]=img
                frame[44:44+633,808:808+414]=imgmodelist[modetype]
                if not ret:
                    print("Error: Could not read frame.")
                    break

                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

                for (x, y, w, h) in faces:
                    roi_gray = gray[y:y+h, x:x+w]
                    label, confidence = recognizer.predict(roi_gray)
                    Roll = label_map[label]
                    
                    curr = time.time()
                    time_string =list[Roll]['Attendance Time']
                    struct_time = time.strptime(time_string, "%a %b %d %H:%M:%S %Y")
                    prev_seconds = time.mktime(struct_time)
                    
                    if curr-prev_seconds>=86400 and 10<counter<=20:
                        student_ref = ref.child(Roll)
                        student_ref.update({
                            'Attendance Time': time.ctime(curr)
                        })
                        modetype=2
                        frame[44:44+633,808:808+414]=imgmodelist[modetype]
                        counter=0
                        
                    elif curr-prev_seconds<86400 and 10<counter<=20:
                        modetype=3
                        frame[44:44+633,808:808+414]=imgmodelist[modetype]
                        counter=0
                        
                    name=list[Roll]['Name']
                    cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                    cv2.rectangle(frame,(x,y-50),(x+w,y),(0,255,0),-1)
                    cv2.putText(frame, "Name: "+name, (x, y-30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
                    cv2.putText(frame, "Roll: "+Roll, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
                    
                if counter>20:
                    modetype=0
                    counter=0
                    frame[44:44+633,808:808+414]=imgmodelist[modetype]

                counter+=1
                
                cv2.imshow('Face Recognition', frame)

                # Exit if 'q' is pressed
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break

            cap.release()
            cv2.destroyAllWindows()
        
        recognize_faces_from_webcam()
        
        

class ProgressBar:
    def __init__(self, root):
        self.root = root
        self.progress_window = ck.CTkToplevel(self.root)
        self.progress_window.geometry("250x150")
        self.progress_window.title("Progress")
        self.progress_window.configure(fg_color="#080707")

        self.progress_bar = ck.CTkProgressBar(master=self.progress_window, width=200, height=25)
        self.progress_bar.pack(pady=50)

        load_label = ck.CTkLabel(master=self.progress_window, text="Loading....")
        load_label.place(x=100, y=90)

        self.update_progress()

    def update_progress(self, step=0):
        if step <= 100:
            self.progress_bar.set(step / 100)
            self.progress_window.after(100, self.update_progress, step + 1)
        else:
            self.progress_window.destroy()
            MessagePopup(self.root)

class MessagePopup:
    def __init__(self, root):
        self.root = root
        self.popup_window = ck.CTkToplevel(self.root)
        self.popup_window.geometry("250x150")
        self.popup_window.title("Successful Registration")
        self.popup_window.configure(fg_color="#080707")

        tick_label = ck.CTkLabel(master=self.popup_window, text="\u2714", font=("Arial", 24),fg_color="green")
        tick_label.place(x=110, y=20)

        success_label = ck.CTkLabel(master=self.popup_window, text="Student Successfully Registered !!")
        success_label.place(x=30, y=50)

        ok_button = ck.CTkButton(master=self.popup_window, text="OK", command=self.popup_window.destroy)
        ok_button.place(x=60, y=100)

        
class Student_register_window:
    def __init__(self,root1):
        
        ck.set_appearance_mode("dark")
        ck.set_default_color_theme("green")
        
        if not firebase_admin._apps:
            cred = credentials.Certificate("serviceaccountkey.json")
            firebase_admin.initialize_app(cred, {
                'databaseURL': "https://faceattendance-63040-default-rtdb.firebaseio.com/"
            })
        root1.withdraw()
        root1.destroy()

        self.root = ck.CTk()
        self.root.geometry("809x500")
        self.root.title("STUDENT REGISTRATION WINDOW")

        #LOGIN IMG
        self.login_image = ck.CTkImage(dark_image=Image.open('Welcome.png'), size=(809, 500)) 
        self.login_label = ck.CTkLabel(master=self.root, text="", image=self.login_image)
        self.login_label.place(x=0, y=0)
        
        self.login_image2 = ck.CTkImage(dark_image=Image.open('Welcome.png'), size=(400, 450)) 
        self.login_label2 = ck.CTkLabel(master=self.root, text="", image=self.login_image2)
        self.login_label2.place(x=20, y=25)

        self.frame = ck.CTkFrame(master=self.root, width=400, height=450, fg_color="#080707",
                                 corner_radius=15, border_color="white", border_width=0.6)
        self.frame.place(x=400, y=25)
        self.studentreg_label = ck.CTkLabel(master=self.frame, text="ð—¦ð—§ð—¨ð——ð—˜ð—¡ð—§  ð—¥ð—˜ð—šð—œð—¦ð—§ð—¥ð—”ð—§ð—œð—¢ð—¡", font=("sans-serif", 30))
        self.studentreg_label.place(x=75, y=10)

        self.name_label = ck.CTkLabel(master=self.frame, text="Student Name :")
        self.name_label.place(x=20, y=80)
        self.name_entry = ck.CTkEntry(master=self.frame, placeholder_text="STUDENT NAME", width=250)
        self.name_entry.place(x=120, y=80)

        self.roll_label = ck.CTkLabel(master=self.frame, text="Roll No. :")
        self.roll_label.place(x=20, y=130)
        self.roll_entry = ck.CTkEntry(master=self.frame, placeholder_text="ROLL NO.", width=250)
        self.roll_entry.place(x=120, y=130)

        self.branch_var = tk.StringVar(value="None")
        self.branch_label = ck.CTkLabel(master=self.frame, text="Branch :")
        self.branch_label.place(x=20, y=180)
        self.branch_menu = ck.CTkOptionMenu(master=self.frame, variable=self.branch_var, values=[
            "Civil Engineering",
            "Mechanical Engineering",
            "Electrical Engineering",
            "Electronics and Communication Engineering",
            "Computer Science Engineering",
            "Chemical Engineering",
            "Aerospace Engineering",
            "Marine Engineering",
            "Biotechnology Engineering",
            "Information Technology",
            "Petroleum Engineering",
            "Nuclear Engineering",
            "Agricultural Engineering",
            # Add more branches as needed
        ], width=250)
        self.branch_menu.place(x=120, y=180)

        self.year_var = tk.StringVar(value="None")
        self.year_label = ck.CTkLabel(master=self.frame, text="Year :")
        self.year_label.place(x=20, y=230)
        self.year1_radio = ck.CTkRadioButton(master=self.frame, text="1st", variable=self.year_var, value="1st")
        self.year1_radio.place(x=110, y=230)
        self.year2_radio = ck.CTkRadioButton(master=self.frame, text="2nd", variable=self.year_var, value="2nd")
        self.year2_radio.place(x=170, y=230)
        self.year3_radio = ck.CTkRadioButton(master=self.frame, text="3rd", variable=self.year_var, value="3rd")
        self.year3_radio.place(x=230, y=230)
        self.year4_radio = ck.CTkRadioButton(master=self.frame, text="4th", variable=self.year_var, value="4th")
        self.year4_radio.place(x=290, y=230)

        self.genderVar = tk.StringVar(value="Prefer not to say")
        self.gender_label = ck.CTkLabel(master=self.frame, text="Gender :")
        self.gender_label.place(x=20, y=280)
        self.gender_male = ck.CTkRadioButton(master=self.frame, text="Male", variable=self.genderVar, value="Male")
        self.gender_male.place(x=120, y=280)
        self.gender_female = ck.CTkRadioButton(master=self.frame, text="Female", variable=self.genderVar, value="Female")
        self.gender_female.place(x=200, y=280)

        self.hall_var=tk.StringVar(value="None")
        self.hall_label=ck.CTkLabel(master=self.frame,text="Hall :")
        self.hall_label.place(x=20,y=330)
        self.hall_menu=ck.CTkOptionMenu(master=self.frame,width=250,variable=self.hall_var,values=[
            "Azad Hall",
            "Birsa Munda Hall (BRH Hall)",
            "Hall of Residence Block Two (Savitribai Phule Hall)",
            "Jagdish Chandra Bose Hall (JCB Hall)",
            "Ladies Hall (LLR Hall)",
            "LBS Hall",
            "Meghnad Saha Hall (MS Hall)",
            "Nehru Hall",
            "Patel Hall",
            "Radhakrishnan Hall (RK Hall)",
            "Ranbir and Chitra Gupta School of Infrastructure Design and Management (RPH Hall)",
            "Sarojini Naidu Hall / Indira Gandhi Hall (S.S.N./I.G. Hall)",
            "Sister Nivedita Hall",
            "Vidyasagar Hall (VS Hall)"
        ])
        self.hall_menu.place(x=120,y=330)

        self.submit_button=ck.CTkButton(master=self.frame,text="SUBMIT",height=35,command=self.handle_submit)
        self.submit_button.place(x=150,y=390)
        
        self.main_button=ck.CTkButton(master=self.root,text="Main Window",height=35,command=self.call,fg_color="red")
        self.main_button.place(x=100,y=420)

        self.root.mainloop()
        
    def call(self):
        Main_Window(self.root,ck.CTk())
        
    def handle_submit(self):
        ref = db.reference('Students')

        roll = f"{self.roll_entry.get()}"
        name = f"{self.name_entry.get()}"
        branch = f"{self.branch_menu.get()}"
        hall = f"{self.hall_menu.get()}"
        gender = f"{self.genderVar.get()}"
        year = f"{self.year_var.get()}"
        curr = time.time()
        Attendance_Time = time.ctime(curr)

        data = {
            roll: {
                "Name": name,
                "Roll": roll,
                "Branch": branch,
                "Hall": hall,
                "Gender": gender,
                "Year": year,
                "Attendance Time": Attendance_Time
            }
        }

        for key, value in data.items():
            ref.child(key).set(value)
            
       

        ProgressBar(self.root)
        
        
class Main_Window:
    def __init__(self,root1,root2):
        
        root1.withdraw()
        root2.withdraw()
        root1.destroy()
        root2.destroy()
        
        self.window = tk.Tk()
        self.window.geometry('1166x718')
        self.window.resizable(0, 0)
        self.window.state('zoomed')
        self.window.title('FACE ATTENDANCE SYSTEM')
        
        self.bg_frame = Image.open('face.jpg')
        self.bg_frame=self.bg_frame.resize((1000,750))
        photo = ImageTk.PhotoImage(self.bg_frame)
        self.bg_panel = Label(self.window, image=photo)
        self.bg_panel.image = photo
        self.bg_panel.place(x=-100,y=0)
        
        heading=Label(self.window,text="Face Attendance System",fg="purple", font=("Trebuchet Ms",35,"bold"))
        heading.place(x=750,y=50)
        
        reg_command=lambda:self.call(self.window)
        reg_button=Button(master=self.window,text="Student Registration",bg='#27b882', fg='white',cursor='hand2', activebackground='#27b882'
                         ,font=("yu gothic ui", 13, "bold"),width=25,command=reg_command)
        reg_button.place(x=880,y=250)
        
        atten_button=Button(master=self.window,text="Face Attendance",bg='#27b882', fg='white',cursor='hand2', activebackground='#27b882'
                         ,font=("yu gothic ui", 13, "bold"),width=25,command=self.call2face)
        atten_button.place(x=880,y=350)
        
        sheet_button=Button(master=self.window,text="Attendance Register",bg='#27b882', fg='white',cursor='hand2', activebackground='#27b882'
                         ,font=("yu gothic ui", 13, "bold"),width=25,command=self.attendance_sheet)
        sheet_button.place(x=880,y=450)
        
        exit_button=Button(master=self.window,text="EXIT",bg='red', fg='white',cursor='hand2', activebackground='red'
                         ,font=("yu gothic ui", 13, "bold"),width=15, command=self.exit_window)
        exit_button.place(x=930,y=650)
        
    
    def exit_window(self):
        result=messagebox.askyesno("Exit","Do you want to exit?")
        if result:
            self.window.withdraw()
            self.window.destroy()
        else:pass
        
    def attendance_sheet(self):
        self.root = tk.Toplevel(self.window)
        self.root.title("Attendance Register")
        self.root.geometry('1166x718')
        
        # Create a Treeview widget for the table
        self.table = ttk.Treeview(self.root)
        self.table["columns"] = ("Roll", "Name", "Branch", "Gender", "Hall", "Year", "Attendance Time")
#         self.table.heading("#0", text="ID")
#         self.table.column("#0", width=30,anchor="w")
        self.table.heading("Roll", text="Roll")
        self.table.column("Roll", width=100,anchor="w")
        self.table.heading("Name", text="Name")
        self.table.column("Name", width=150,anchor="w")
        self.table.heading("Branch", text="Branch")
        self.table.column("Branch", width=200,anchor="w")
        self.table.heading("Gender", text="Gender")
        self.table.column("Gender", width=100,anchor="w")
        self.table.heading("Hall", text="Hall")
        self.table.column("Hall", width=100,anchor="w")
        self.table.heading("Year", text="Year")
        self.table.column("Year", width=100,anchor="w")
        self.table.heading("Attendance Time", text="Attendance Time")
        self.table.column("Attendance Time", width=200,anchor="w")
        # Add sample data to the table (you can replace this with your actual data)
        self.add_data_to_table()

        # Pack the table
        self.table.pack(fill=tk.BOTH, expand=True)

    def add_data_to_table(self):
        if not firebase_admin._apps:
            cred = credentials.Certificate("serviceaccountkey.json")
            firebase_admin.initialize_app(cred, {
                'databaseURL': "https://faceattendance-63040-default-rtdb.firebaseio.com/"
            })
            
        ref=db.reference('Students')
        # Insert sample data into the table
        list=ref.get()
        id=1
        for roll, info in list.items():
            if roll=='Harsh Dalmia':
                pass
            else:
                values = (
                    roll,
                    info.get('Name', ''),
                    info.get('Branch', ''),
                    info.get('Gender', ''),
                    info.get('Hall', ''),
                    info.get('Year', ''),
                    info.get('Attendance Time', '')
                )
                self.table.insert("", "end", values=values)
                id=id+1
    
    def call(self,root):
        Student_register_window(root)
        
    def call2face(self):
        face_recog(self.window)
        
      
    
    

class LoginPage:
    def __init__(self, window):
        window.withdraw()
        window.destroy()
        self.window =tk.Tk()
        self.window.geometry('1166x718')
        self.window.resizable(0, 0)
        self.window.state('zoomed')
        self.window.title('Login Page')
        
        if not firebase_admin._apps:
            cred = credentials.Certificate("serviceaccountkey.json")
            firebase_admin.initialize_app(cred, {
                'databaseURL': "https://faceattendance-63040-default-rtdb.firebaseio.com/"
            })

        # ========================================================================
        # ============================background image============================
        # ========================================================================
        self.bg_frame = Image.open('background1.png')
        photo = ImageTk.PhotoImage(self.bg_frame)
        self.bg_panel = Label(self.window, image=photo)
        self.bg_panel.image = photo
        self.bg_panel.pack(fill='both', expand='yes')
        # ====== Login Frame =========================
        self.lgn_frame = Frame(self.window, bg='#040405', width=950, height=600)
        self.lgn_frame.place(x=200, y=70)

        # ========================================================================
        # ========================================================
        # ========================================================================
        self.txt = "WELCOME"
        self.heading = Label(self.lgn_frame, text=self.txt, font=('yu gothic ui', 25, "bold"), bg="#040405",
                             fg='white',
                             bd=5,
                             relief=FLAT)
        self.heading.place(x=80, y=30, width=300, height=30)

        # ========================================================================
        # ============ Left Side Image ================================================
        # ========================================================================
        self.side_image = Image.open('vector.png')
        photo = ImageTk.PhotoImage(self.side_image)
        self.side_image_label = Label(self.lgn_frame, image=photo, bg='#040405')
        self.side_image_label.image = photo
        self.side_image_label.place(x=5, y=100)

        # ========================================================================
        # ============ Sign In Image =============================================
        # ========================================================================
        self.sign_in_image = Image.open('hyy.png')
        photo = ImageTk.PhotoImage(self.sign_in_image)
        self.sign_in_image_label = Label(self.lgn_frame, image=photo, bg='#040405')
        self.sign_in_image_label.image = photo
        self.sign_in_image_label.place(x=620, y=130)

        # ========================================================================
        # ============ Sign In label =============================================
        # ========================================================================
        self.sign_in_label = Label(self.lgn_frame, text="Sign In", bg="#040405", fg="white",
                                    font=("yu gothic ui", 17, "bold"))
        self.sign_in_label.place(x=650, y=240)

        # ========================================================================
        # ============================username====================================
        # ========================================================================
        self.username_label = Label(self.lgn_frame, text="Username", bg="#040405", fg="#4f4e4d",
                                    font=("yu gothic ui", 13, "bold"))
        self.username_label.place(x=550, y=300)

        self.username_entry = Entry(self.lgn_frame, highlightthickness=0, relief=FLAT, bg="#040405", fg="#6b6a69",
                                    font=("yu gothic ui ", 12, "bold"), insertbackground = '#6b6a69')
        self.username_entry.place(x=580, y=335, width=270)

        self.username_line = Canvas(self.lgn_frame, width=300, height=2.0, bg="#bdb9b1", highlightthickness=0)
        self.username_line.place(x=550, y=359)
        # ===== Username icon =========
        self.username_icon = Image.open('username_icon.png')
        photo = ImageTk.PhotoImage(self.username_icon)
        self.username_icon_label = Label(self.lgn_frame, image=photo, bg='#040405')
        self.username_icon_label.image = photo
        self.username_icon_label.place(x=550, y=332)

        # ========================================================================
        # ============================login button================================
        # ========================================================================
        self.lgn_button = Image.open('btn1.png')
        photo = ImageTk.PhotoImage(self.lgn_button)
        self.lgn_button_label = Label(self.lgn_frame, image=photo, bg='#040405')
        self.lgn_button_label.image = photo
        self.lgn_button_label.place(x=550, y=450)
        self.login = Button(self.lgn_button_label, text='LOGIN', font=("yu gothic ui", 13, "bold"), width=25, bd=0,
                            bg='#3047ff', cursor='hand2', activebackground='#3047ff', fg='white',
                            command=self.login_sucessfull)
        self.login.place(x=20, y=10)
        # ========================================================================
        # ============================Forgot password=============================
        # ========================================================================
        self.forgot_button = Button(self.lgn_frame, text="Forgot Password ?",
                                    font=("yu gothic ui", 13, "bold underline"), fg="white", relief=FLAT,
                                    activebackground="#040405"
                                    , borderwidth=0, background="#040405", cursor="hand2")
        self.forgot_button.place(x=630, y=510)
        # =========== Sign Up ==================================================
        self.sign_label = Label(self.lgn_frame, text='No account yet?', font=("yu gothic ui", 11, "bold"),
                                relief=FLAT, borderwidth=0, background="#040405", fg='white')
        self.sign_label.place(x=550, y=560)

        self.signup_img = ImageTk.PhotoImage(file='register.png')
        self.signup_button_label = Button(self.lgn_frame, image=self.signup_img, bg='#98a65d', cursor="hand2",
                                          borderwidth=0, background="#040405", activebackground="#040405")
        self.signup_button_label.place(x=670, y=555, width=111, height=35)

        # ========================================================================
        # ============================password====================================
        # ========================================================================
        self.password_label = Label(self.lgn_frame, text="Password", bg="#040405", fg="#4f4e4d",
                                    font=("yu gothic ui", 13, "bold"))
        self.password_label.place(x=550, y=380)

        self.password_entry = Entry(self.lgn_frame, highlightthickness=0, relief=FLAT, bg="#040405", fg="#6b6a69",
                                    font=("yu gothic ui", 12, "bold"), show="*", insertbackground = '#6b6a69')
        self.password_entry.place(x=580, y=416, width=244)

        self.password_line = Canvas(self.lgn_frame, width=300, height=2.0, bg="#bdb9b1", highlightthickness=0)
        self.password_line.place(x=550, y=440)
        # ======== Password icon ================
        self.password_icon = Image.open('password_icon.png')
        photo = ImageTk.PhotoImage(self.password_icon)
        self.password_icon_label = Label(self.lgn_frame, image=photo, bg='#040405')
        self.password_icon_label.image = photo
        self.password_icon_label.place(x=550, y=414)
        # ========= show/hide password ==================================================================
        self.show_image = ImageTk.PhotoImage \
            (file='show.png')

        self.hide_image = ImageTk.PhotoImage \
            (file='hide.png')

        self.show_button = Button(self.lgn_frame, image=self.show_image, command=self.show, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        self.show_button.place(x=860, y=420)

    def show(self):
        self.hide_button = Button(self.lgn_frame, image=self.hide_image, command=self.hide, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        self.hide_button.place(x=860, y=420)
        self.password_entry.config(show='')

    def hide(self):
        self.show_button = Button(self.lgn_frame, image=self.show_image, command=self.show, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        self.show_button.place(x=860, y=420)
        self.password_entry.config(show='*')
        
    def call(self,root1,root2,check):
        if check==1:
            Main_Window(root1,root2)
        else:
            root1.withdraw()
            root1.destroy()
        
    def login_message(self, message, symbol):
        ck.set_appearance_mode("dark")
        ck.set_default_color_theme("green")
        root = ck.CTk()
        root.title("Login Status")   
        root.configure(fg_color="black")
        root.geometry("250x150")
        
        if symbol==0:
            cross_label=ck.CTkLabel(root,text="\u274C",fg_color="red",font=("Arial", 17))
            cross_label.place(x=110,y=20)
        else:
            tick_label=ck.CTkLabel(root,text="\u2714",fg_color="green",font=("Arial", 17))
            tick_label.place(x=100,y=20)
            
        message_label = ck.CTkLabel(root, text=message, fg_color="black",font=("Arial", 15))
        message_label.pack(padx=10, pady=60)
        
        ok_command=lambda:self.call(root,self.window,symbol)
        
        ok_b=ck.CTkButton(master=root,text="OK",command=ok_command,width=40,height=25)
        ok_b.place(x=100,y=100)
        root.mainloop()

        
    def login_sucessfull(self):
        
        ref=db.reference('Login')
        username=f"{self.username_entry.get()}"
        password=f"{self.password_entry.get()}"
        
        list=ref.get()
        
        actual_username=list['Username']
        actual_password=list['Password']
        success=0
        if username==actual_username and password==actual_password :
            self.login_message('Login Successfull!!',1)
            success=1
        else:
            self.login_message('Login Failed, Try Again!',0)
            
            
            
class Welcome_Window:
    def __init__(self,window):
        self.window = window
        self.window.geometry('1166x718')
        self.window.resizable(0, 0)
        self.window.state('zoomed')
        self.window.title('WELCOME TO FACE ATTENDANCE SYSTEM')
        
        self.bg_frame = Image.open('face3.jpg')
        photo = ImageTk.PhotoImage(self.bg_frame)
        self.bg_panel = Label(self.window, image=photo,text="Face Attendance System")
        self.bg_panel.image = photo
        self.bg_panel.pack(fill='both', expand='yes')
        
#         style = ttk.Style(self.window)
#         style.theme_use('clan')
#         style.configure("TProgressbar", troughcolor='#D3D3D3', background='#40E0D0', thickness=20)
        
        self.progress_bar = ttk.Progressbar(self.window, orient='horizontal', mode='determinate', length=500)
        self.progress_bar.place(x=450,y=630)

        self.load_label = Label(master=self.window, text="Loading....",bg="black",fg="white", font=("Trebuchet Ms",13,"bold"))
        self.load_label.place(x=600,y=660)
    
                
        self.update_progress()
        
    def update_progress(self, step=0):
            if step <= 100:
                self.progress_bar['value'] = step
                self.window.after(100, self.update_progress, step + 1)
                txt="Loading...."+str(step)+"%"
                self.load_label.configure(text=txt)
            else:
                LoginPage(self.window)
      
    

def page():
    window = tk.Tk()
    Welcome_Window(window)
    window.mainloop()
        
     

```
